```python
import numpy as np

n = 2000
x = np.linspace(0, 4*np.pi, n)
y = np.linspace(0, 4*np.pi, n)
X, Y = np.meshgrid(x, y)
matrix = (np.sin(X) * np.cos(2*Y) +
                np.tanh(0.3*X) * np.sin(0.7*Y) +
                0.4 * np.cos(3*X + Y) +
                0.2 * np.random.randn(n, n) + 0.5)
np.fill_diagonal(matrix, matrix.diagonal() + 1.0)
matrix[-1, -1] = 3.14159

print(f"size: {matrix.shape}")
print(f"elements: {matrix.size}")
```
